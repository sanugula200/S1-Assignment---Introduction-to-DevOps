# Introduction to Devops

1. What is DevOps? Explain its goals and benefits.
  
    -> DevOps a blend/combination of two words "Development" and other is "Operations" is a cultural philosophy and set of technical practices that unifies IT, software development, and operations teams. It breaks down traditional silos to automate and optimize the software delivery lifecycle. The Devops tutorial will help you to learn DevOps basics and provide depth knowledge of various DevOps tools such as Git, Ansibke, Docker, Puppet, Jenkins, Chef, Nagios, and Kubernetes to deliver high-quality software safely.

    **Goals of DevOps:**
    
    The main target of DevOps is to make software creation smoother, faster, and error-free. Its goals include:
    
    Fast Delivery: Get new features, updates, or bug fixes out to real users as quickly as possible.
    
    Constant Teamwork: Break down the walls between departments so coding teams and server management teams share the same goals.
    
    Automate Everything: Eliminate boring, repetitive manual work (like manually testing software) by letting smart computer scripts handle it.
    
    Continuous Feedback: Listen to how the software is running in the real world and immediately use that data to plan the next upgrade.

    **Benefits of DevOps:**

    When a company successfully switches to a DevOps mindset, they gain huge advantages across their business:
    
    1. Incredible SpeedInstead of waiting six months to launch one massive, scary software update, companies can launch small, safe updates multiple times a day. If a competitor launches a new feature, a DevOps team can respond and update their own app in a matter of hours.
    
    2. High Quality and Fewer CrashesBecause human beings make mistakes, DevOps uses automated testing assembly lines (called CI/CD pipelines). Every single time a developer writes a line of code, computers automatically check it for bugs. If a bug is found, it is stopped immediately before it ever reaches the customer.
    
    3. Lightning-Fast RecoveryEven with great testing, mistakes can happen. If a new update goes live and accidentally crashes the app, DevOps tools can instantly "undo" the update. The system automatically rolls back to the previous working version in seconds, keeping customers happy while engineers fix the issue in the background.
    
    4. Identical EnvironmentsIn the past, developers would say, "Well, the app works fine on my laptop!" only for it to crash on the company's main servers. DevOps uses tools to write down infrastructure setups as simple text files. This means the developer's laptop, the testing system, and the live public website are 100% identical copies, eliminating surprise errors.
    
    5. Happier, Less Stressed TeamsEngineers no longer have to stay up until 3:00 AM on weekends to launch stressful software updates manually. Since automation handles the heavy lifting, teams are less burned out and can spend their time inventing creative new features.

2. Explain the history and evolution of DevOps.
  
    -> DevOps evolved to break down the traditional silos between software developers and IT operations, shifting software delivery from slow, risky releases to rapid, continuous deployments.
    1. The Problem:   The Pre-DevOps Era (Before 2008)Waterfall Model: Software was built in slow, isolated, linear phases. Siloed Teams: Developers wrote code; operations teams deployed and maintained it. Wall of Confusion: Developers wanted fast changes, while operations demanded system stability.Slow Releases: Deployments happened quarterly or yearly, resulting in massive, buggy updates.
    2. The Spark: Agile and Early Automation (2001–2008) Agile Manifesto (2001): Speed up development cycles, but left operations teams struggling to keep pace. Virtualization: VMware and early cloud services made infrastructure faster to spin up. Configuration Management: Tools like Puppet (2005) introduced infrastructure managed by code.
    3. The Birth of DevOps (2009) Flickr’s Presentation: Engineers proved collaboration allowed them to run 10 deployments a day.The Term Coined: Patrick Debois organized the first "DevOpsDays" conference in Belgium. The Portmanteau: "Development" and "Operations" merged into the single term DevOps.
    4. The Tooling Explosion (2010–2015)CI/CD Pipelines: Jenkins automated code testing and continuous delivery. Containers: Docker (2013) standardized how software is packaged and run. Orchestration: Kubernetes (2014) emerged to automate and manage container deployment.
    5. Enterprise & Modern DevOps (2016–Present) DevSecOps: Security shifted left, becoming integrated directly into the deployment pipeline. Platform Engineering: Internal developer platforms were built to reduce developer workload.AI & GitOps: AI automates code reviews, while Git handles automated infrastructure changes

3. What is SDLC? Explain traditional development models.
  
    -> SDLC (Software Development Life Cycle) is a step by step process used by engineering teams to plan, design, build, test, and deploy high-quality software efficiently. Traditional Development ModelsTraditional models are highly structured, linear frameworks that focus on strict planning and documentation.
    
    Waterfall Model: A rigid, step by step approach where each phase must finish completely before the next begins.
    
    V-Model: A variation of Waterfall where a specific testing phase is planned directly alongside every development phase.
    
    Iterative & Incremental: Software is built in small, repeated sections, adding features with each new version.
    
    Spiral Model: A risk-driven approach that repeatedly cycles through planning, risk analysis, and building.

4. Explain Agile methodologies: Scrum and Kanban.

    -> Agile methodologies break software development into fast, iterative cycles to help teams adapt quickly to changing user demands. The two most popular frameworks used to implement Agile are Scrum and Kanban.
    1. Scrum (Structure & Sprints)Scrum is a highly structured framework that organizes work into fixed, short timeblocks.
    
        Sprints: Work happens in fixed-length cycles, usually lasting 2 to 4 weeks.
    
        Roles: Teams include a Product Owner (manages goals), a Scrum Master (removes obstacles), and Developers.
    
        Ceremonies: Teams hold daily standups, sprint planning, sprint reviews, and retrospectives.
    
        Artifacts: Progress is tracked via a Product Backlog, Sprint Backlog, and an Increment (working software).
    
    2. Kanban (Flexibility & Flow)Kanban is a continuous, highly visual framework focused on managing steady workflow without fixed deadlines.
    
        Kanban Board: Work is visually tracked across columns like To Do, In Progress, and Done.
        
        WIP Limits: Teams set "Work-in-Progress" limits to prevent multitasking and bottlenecking.
    
        Continuous Delivery: Software is deployed continuously as soon as tasks are finished.
        
        Key Metrics: Success is measured by Lead Time (total time from start to finish) and Cycle Time.
        
        Key Difference Scrum relies on rigid timeboxes (Sprints) and predefined team roles, while Kanban focuses on a continuous flow of tasks with maximum flexibility.

5. Explain the role of an Operating System and the difference
between processes and threads, process lifecycle.

    -> An Operating System (OS) is the core software that manages hardware components and provides a secure platform for applications to run. It acts as an intermediary, handling memory allocation, file systems, device communication, and CPU scheduling.
    
    Here is a balanced overview of processes, threads, and the process lifecycle.Processes vs. ThreadsThe difference lies in how they utilize system memory and resources.
    
    Process: An independent program in execution (e.g., opening a web browser). Each process has its own isolated memory space. It has high overhead to create, but if one process crashes, it does not affect others.
    
    Thread: A lightweight execution path inside a process (e.g., a single tab or a spell-checker inside that browser). Threads share the parent process’s memory. They switch incredibly fast, but if one thread crashes, it can crash the entire process.
    
    **The Process Lifecycle**
    
    The OS uses a state machine to manage a process from creation to deletion as it competes for CPU time:
    
    New: The process is being created and loaded from storage into RAM.
    
    Ready: The process is fully loaded in memory and waiting in a queue for the CPU scheduler to run it.
    
    Running: The CPU is actively executing the instructions of the process.
    
    Waiting (Blocked): The process is temporarily paused because it is waiting for an external event, such as a file loading from a disk or user input. Once done, it returns to the Ready state.
    
    Terminated: The process finishes its task or is closed. The OS clears it from memory and reclaims its resources.
